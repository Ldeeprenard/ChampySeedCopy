```python
import pandas as pd
```

importation du tableau principal et une liste créée avec la comestibilité des champignons

```python
df = pd.read_csv ("tableau_propre.csv")
comestible = pd.read_excel ("liste comestible.xlsx")
```

```python
newtab = pd.merge(df, comestible, left_on='verbatimScientificName', right_on='champignons', how="left")
```

```python
newtab.columns
```

```python
newtab = newtab.dropna (subset = ["comestibilité"],axis=0)
```

```python
newtab.shape
```

```python
newtab["comestibilité"].value_counts().plot.bar(legend="comestibilité")
```

```python
newtab["family"].value_counts().plot.bar(legend='family')
```
