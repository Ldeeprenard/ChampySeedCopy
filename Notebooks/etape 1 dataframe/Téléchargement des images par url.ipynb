## Téléchargement des images quand on a les URL
On peut au choix les télécharger dans un logiciel externe (mais ça fait beaucoup de liens à copier)  
ou bien faire des requêtes sur les lignes de notre dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C:/Users/bhozj/Documents/Datascientest/ProjetMushrooms/tableau_propre.csv")
df.head()
```

Ce notebook sert à télécharger N fichiers aléatoirement à des fins exploratoires, changer le nombre de samples pour modifier le nombre de fichiers téléchargés

```python
#liste d'url contenant N fichiers choisis au hasard 
df_list_url = df["identifier_y"].sample(10)
```

```python
import requests

for i in range(len(df_list_url)):
    print(df_list_url.iloc[i])
    url = df_list_url.iloc[i]

    r = requests.get(url)
    with open("C:/Users/bhozj/Documents/Datascientest/ProjetMushrooms/Images/"+str(i)+".jpg", "wb") as f:
        f.write(r.content)    
    
```
