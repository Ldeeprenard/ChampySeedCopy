```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv ("Comestible20210903.csv")
```

```python
df.shape
```

```python
df.identifier_y.unique().shape
```

```python
familycount = df["family"].value_counts()
```

affichage du nombre de photos par famille

```python
sns.countplot (x="family",data=df)
```

```python
bins = [1,100,1000,5000,10000,15000,20000,30000,40000,50000,60000,100000]
```

```python
test = pd.cut (familycount,bins)
```

affichage du nombre de famille par tranche de quantité de photos

```python
plt.figure (figsize=[20,10])
ax = sns.countplot (x=test);
ax.set(xlabel="Nombre de photos", ylabel="Nombre de familles", title='nb de familles/qte de photos');
```
