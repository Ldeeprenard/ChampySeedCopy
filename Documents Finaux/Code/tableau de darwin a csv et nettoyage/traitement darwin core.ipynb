Création du tableau en mergeant à droite (ne créer les lignes qu'en fonction des photos)

```python
tableau_final = pd.merge(occu_df, img_df, left_on='id', right_on='coreid', how="right")
tableau_final = tableau_final.dropna (axis=1, thresh= len(tableau_final)*0.7)
```
